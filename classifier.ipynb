### Classifier for image data
Splits location of images into clusters

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import pickle
```

```python
raw = pd.read_csv('train.csv')
coords = raw[['x', 'y']]
coords = coords.drop_duplicates(subset = None)
```

```python
gmm = GaussianMixture(n_components = 8)
gmm.fit(coords)
```

```python
classes = gmm.predict(coords)
plt.scatter(coords.x, coords.y, c = classes)
```

```python
with open('gmm.pkl', 'wb') as f:
    pickle.dump(gmm, f)
```
