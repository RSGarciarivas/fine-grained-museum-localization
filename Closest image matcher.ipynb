### Pick a random image with features from the dataset

```python
# Get all image paths from test set
img_paths = glob.glob('./Dataset/test/*')

# Initialise SIFT detector
sift = cv.SIFT_create()

# Loop until image with enough features is found
while True:
    
    # Get a random image
    r = np.random.randint(0, len(img_paths))
    img_path = img_paths[r]
    img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)
    
    # Check if image has sufficient features
    kp, _ = sift.detectAndCompute(img, None)
    if len(kp) >= 5:
        break
        
# Plot image
plt.subplots(figsize = (15, 15))
plt.imshow(img, cmap = 'gray')
plt.axis('off')
plt.show()
```
