# Splitting the training dataset into subsets for validating on labelled data

```python
import glob
import numpy as np
import pickle
import csv

from utils import SIFT_feature_dataset, save_dataset
```

### List all images in the training set

```python
# Get all images in directory
img_paths = glob.glob('./Dataset/train/*')
```

### Split list of paths into train and validation

```python
train_ratio = 0.85 # Number between 0 and 1

# Vector of random numbers from 0 to 1 (same size as dataset)
R = np.random.rand(len(img_paths))

# Arrays for storing image paths for subsets
validation_paths = []
training_paths = []

# Go through each image path and assign it to a set
for i, path in enumerate(img_paths):
    if R[i] < train_ratio:
        training_paths.append(path)
        
    else:
        validation_paths.append(path)

print(f'Training set: {len(training_paths)} images')
print(f'Validation set: {len(validation_paths)} images')
```

### Create training subset

```python
training_subset = SIFT_feature_dataset(training_paths)
```

### Save subset and list of validation image paths

```python
save_dataset('training_subset', training_subset)

with open('validation_paths' + '.pkl', 'wb') as output_file:
    pickle.dump(validation_paths, output_file)
```

### Create and save dictionary with all image labels

```python
labels = {}

with open('./Dataset/train.csv') as labels_csv:
    reader = csv.reader(labels_csv)
    # Ignore header
    next(reader)
    
    # Read each row
    for row in reader:
        labels[row[0]] = (float(row[1]), float(row[2]))
        
with open('image_labels.pkl', 'wb') as output_file:
    pickle.dump(labels, output_file)
```
