# Build a dictionary containing SIFT keypoints and descriptors for each image in the training set, save it as .pkl file

### Necessary imports

```python
import glob
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import pickle

from utils import SIFT_feature_dataset, get_img_features, save_dataset
```

### Build SIFT features dictionary

```python
# Get list of all images
SIFT_features = SIFT_feature_dataset('./Dataset/train/*')
```

### Visualise random image with features

```python
# Display random image
img_paths = glob.glob('./Dataset/train/*')
i = np.random.randint(0, len(SIFT_features))
img_path = img_paths[i]

# Load image
img = cv.cvtColor(cv.imread(img_path), cv.COLOR_BGR2RGB)
img_name = img_path[16:-4]

# Load keypoints and descriptors
kp, des = get_img_features(img_name, SIFT_features)

# Draw features on image
kp_img = cv.drawKeypoints(img, kp, None, flags = cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Plot
plt.subplots(figsize = (15, 15))
plt.imshow(kp_img)
plt.axis('off')
plt.show()
```
